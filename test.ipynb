```python
df1 = pd.read_csv("Cleaned_LinkedIn_Job_Postings_Dataset.csv")
df2 = pd.read_csv("linkedin_jobs_sample.csv")
df3 = pd.read_csv("linkedin-scrape_data-scientist_2023-02-25.csv")
df4 = pd.read_csv("linkedin-scrape_data-scientist_2023-03-08.csv")
df5 = pd.read_csv("linkedin-scrape_machine-learning_2023-01-03.csv")
```

```python
# df1 temizleme
df1 = df1.rename(columns={
    "name": "company",
    "title": "position"
})

df1["post_date"] = None  # tarih kolonu yok

df1 = df1[["company", "position", "location", "description", "post_date"]]
```

```python
# df2 temizleme
df2 = df2.rename(columns={
    "title": "position",
    "url": "description"
})

df2["post_date"] = None

df2 = df2[["company", "position", "location", "description", "post_date"]]
```

```python
# df4 temizleme
df4 = df4.rename(columns={"full_text": "description"})

df4["post_date"] = pd.to_datetime(df4["post_date"], errors="coerce")

df4 = df4[["company", "position", "location", "description", "post_date"]]
```

```python
# df5 temizleme
df5 = df5.rename(columns={"full_text": "description"})

df5["post_date"] = pd.to_datetime(df5["post_date"], errors="coerce")

df5 = df5[["company", "position", "location", "description", "post_date"]]
```

```python
# Tüm dataframe'leri tek listede topla
dfs = [df1, df2, df3, df4, df5]

# Dikey olarak birleştir
merged_df = pd.concat(dfs, ignore_index=True)

print(merged_df.head())
print(merged_df.tail())
print(merged_df.shape)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# NLTK içeriklerini bir kez indir
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    if pd.isna(text):
        return ""
    
    # 1) Küçük harf
    text = text.lower()
    
    # 2) URL'leri kaldır
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    
    # 3) Sadece harf kalsın
    text = re.sub(r"[^a-z\s]", " ", text)
    
    # 4) Tokenize
    words = text.split()
    
    # 5) Stopwords temizle
    words = [w for w in words if w not in stop_words]
    
    # 6) Lemmatization
    words = [lemmatizer.lemmatize(w) for w in words]
    
    # 7) Tekrar birleştir
    return " ".join(words)

# Yeni column olarak temizlenmiş description yaz
merged_df["clean_desc"] = merged_df["description"].apply(clean_text)

# İlk 5 satır
merged_df[["description", "clean_desc"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
# Boş metinleri filtreleyelim (çok nadir de olsa olabilir)
text_data = merged_df["clean_desc"].fillna("")

vectorizer = TfidfVectorizer(
    max_df=0.7,        # çok sık geçenleri biraz kırp
    min_df=10,         # 10'dan az geçen kelimeleri at
    max_features=5000, # en çok bilgi veren 5000 özellik
    ngram_range=(1, 2) # unigram + bigram
)

X = vectorizer.fit_transform(text_data)

X.shape
```

```python
k = 5  # şimdilik 5 küme deneyelim

kmeans = KMeans(
    n_clusters=k,
    random_state=42,
    n_init=10
)

kmeans.fit(X)

# Her ilana hangi cluster geldi?
merged_df["cluster"] = kmeans.labels_

merged_df[["position", "location", "cluster"]].head()
```

```python

import numpy as np

terms = vectorizer.get_feature_names_out()

for i in range(k):
    print(f"\n=== Küme {i} ===")
    # O kümenin merkezindeki en yüksek ağırlıklı kelimeleri bul
    center = kmeans.cluster_centers_[i]
    top_idx = np.argsort(center)[::-1][:15]
    top_terms = [terms[j] for j in top_idx]
    print("En baskın kelimeler:", ", ".join(top_terms))
```
